```python
# Creating Categories 
```

```python
def mode_cat(mode):
    if mode == 1:
        return 'Major'
    elif mode == 0:
        return 'Minor'
    else:
        return 'Unknown'
```

```python
df_spty_clean['mode_Cat'] = df_spty_clean['mode'].apply(mode_cat)
```

```python
def key_cat(key):
    if key == 0:
        return 'C'
    elif key == 1:
        return 'C#/Db'
    if key == 2:
        return 'D'
    elif key == 3:
        return 'D#/Eb'
    if key == 4:
        return 'E'
    elif key == 5:
        return 'E#/Fb'
    if key == 6:
        return 'F'
    elif key == 7:
        return 'F#/Gb'
    if key == 8:
        return 'G'
    elif key == 9:
        return 'G#/Bb'
    if key == 10:
        return 'B'
    elif key == 11:
        return 'B#/Cb'
    else:
        return 'Unknown'
```

```python
df_spty_clean['key_cat'] = df_spty_clean['key'].apply(key_cat)
```

```python
# Converting Milliseconds to seconds
```

```python
df_spty_clean['duration_sec'] = df_spty_clean['duration_ms']*1e-3
df_spty_clean = df_spty_clean.drop(['duration_ms'], axis=1)
```
